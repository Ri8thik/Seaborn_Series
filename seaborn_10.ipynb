```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
Gw_df=pd.read_csv("Who_is_responsible_for_global_warming.csv")
Gw_df
```

```python
new_df=Gw_df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
new_df.head(6)
```

```python
new_df.corr()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df.corr(),annot=True,linecolor="w",linewidths=3)
plt.title("correlation of globalworming ",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(new_df.corr(),annot=True,linecolor="w",linewidths=3)
ax.tick_params(size=10,color="k",labelsize=10,labelcolor="k")
plt.title("correlation of globalworming ",fontsize=20)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_data=load_breast_cancer()
cancer_data
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_data["data"],cancer_data["target"]],
                       columns=np.append(cancer_data["feature_names"], ["target"]))
cancer_df
```

```python
plt.figure(figsize=(25,25))
ax=sns.heatmap(cancer_df.corr(),annot=True,linecolor="w",linewidths=3)
ax.tick_params(size=10,color="k",labelsize=10,labelcolor="k")
plt.title("correlation of breast cancer ",fontsize=20)
plt.show()
```
