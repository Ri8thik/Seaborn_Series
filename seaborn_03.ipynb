```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
data_df=sns.load_dataset("tips")
data_df
```

```python
sns.distplot(data_df["total_bill"])
plt.show()
```

```python
sns.distplot(data_df["total_bill"],bins=100,hist=True)
plt.show()
```

```python
sns.distplot(data_df["total_bill"],bins=100,hist=False)
plt.show()
```

```python
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=False)
plt.show()
```

```python
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=True,rug=True,fit=norm)
plt.show()
```

```python
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=True,rug=True,fit=norm,color="g")
plt.show()
```

```python
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=True,rug=True,fit=norm,color="g",vertical=True)
plt.show()
```

```python
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=True,rug=True,fit=norm,color="g",vertical=False,label="Total_bill",
             axlabel="Total_bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=True,rug=True,fit=norm,color="g",vertical=False,label="Total_bill",
             axlabel="Total_bill")
plt.title("The Bill Histogram",fontsize=20)
plt.xlabel("Total bill",fontsize=15)
plt.ylabel("Density",fontsize=15)
plt.legend()
plt.show()
```
