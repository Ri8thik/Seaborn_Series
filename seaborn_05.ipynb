```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_df=sns.load_dataset("tips")
data_df
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x="day",y="total_bill",data=data_df)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x="day",y="total_bill",data=data_df,hue="sex")
plt.show()
```

```python
order=["Fri","Thur" ,"Sun","Sat"] 
plt.figure(figsize=(16,9))
sns.barplot(x="day",y="total_bill",data=data_df,hue="sex",order=order)
plt.show()
```

```python
order=["Fri","Thur" ,"Sun","Sat"] 
plt.figure(figsize=(16,9))
hue_order=["Female","Male"]
sns.barplot(x="day",y="total_bill",data=data_df,hue="sex",order=order,hue_order=hue_order)
plt.show()
```

```python
order=["Fri","Thur" ,"Sun","Sat"] 
plt.figure(figsize=(16,9))
hue_order=["Female","Male"]
sns.barplot(y="day",x="total_bill",data=data_df,hue="sex",order=order,hue_order=hue_order,estimator=np.max,ci=100,orient="h")
plt.show()
```

```python
order=["Fri","Thur" ,"Sun","Sat"] 
plt.figure(figsize=(16,9))
hue_order=["Female","Male"]
sns.barplot(x="day",y="total_bill",data=data_df,hue="sex",order=order,hue_order=hue_order,estimator=np.max,ci=100,
            orient="v",
            color="g",
            palette="hot",
            saturation=0.5,
            errcolor='0.5',
            errwidth=12,
            capsize=0.2)
plt.show()
```

```python
order=["Fri","Thur" ,"Sun","Sat"] 
plt.figure(figsize=(16,9))
hue_order=["Female","Male"]
sns.barplot(x="day",y="total_bill",data=data_df,hue="sex",order=order,hue_order=hue_order,estimator=np.max,ci=100,
            orient="v",
            color="g",
            palette="hot",
            saturation=0.5,
            errcolor='0.5',
            errwidth=12,
            capsize=0.2,dodge=False)
plt.show()
```
