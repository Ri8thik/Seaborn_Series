```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_df=sns.load_dataset("tips")
data_df
```

```python
sns.barplot(x="day",y="total_bill",data=data_df)
```

```python
kwargs={"alpha":0.9,"linewidth":2,"linestyle":"--","edgecolor":"k"}
sns.barplot(x="day",y="total_bill",data=data_df,**kwargs)
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=data_df,alpha=0.9,linewidth=2,linestyle="--",edgecolor="k")
plt.show()
```

```python
ax=sns.barplot(x="day",y="total_bill",data=data_df,alpha=0.9,linewidth=2,linestyle="--",edgecolor="k")
ax.set(title="title",xlabel="day",ylabel="total_bill")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x="day",y="total_bill",data=data_df,alpha=0.9,linewidth=2,linestyle="--",edgecolor="k")
plt.title("title")
plt.xlabel("Days")
plt.ylabel("Total_bill")
plt.show()
```
