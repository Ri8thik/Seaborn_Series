```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data_df=np.linspace(0,15,12).reshape((4,3))
data_df
```

```python
sns.heatmap(data_df)
```

```python
Gw_df=pd.read_csv("Who_is_responsible_for_global_warming.csv")
Gw_df
```

```python
new_df=Gw_df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
new_df.head(6)
```

```python
sns.heatmap(new_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df,vmin=0,vmax=21)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df,vmin=0,vmax=21,cmap="coolwarm")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df,vmin=0,vmax=21,cmap="coolwarm",annot=True)
plt.show()
```
