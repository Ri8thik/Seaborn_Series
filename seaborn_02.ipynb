```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
data_set=sns.load_dataset("tips")
data_set
```

```python
plt.figure(figsize=(16,9))
sns.set(style="darkgrid")
sns.lineplot(x="total_bill",y="tip",data=data_set,hue="sex",style="sex",palette="hot",dashes=False,markers=["o","<"],legend="brief")
plt.title("Tips Data Set",fontsize=20)
plt.xlabel("Total Bill",fontsize=15)
plt.ylabel("TIP",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style="darkgrid")
sns.lineplot(x="total_bill",y="tip",data=data_set,hue="day",style="day",palette="hot",dashes=False,markers=["o","<",".","*"],legend="brief")
plt.title("Tips Data Set",fontsize=20)
plt.xlabel("Total Bill",fontsize=15)
plt.ylabel("TIP",fontsize=15)
plt.show()
```
