```python
import matplotlib.pyplot as plt
import seaborn as sns 
from scipy.stats import norm
```

```python
data_df=sns.load_dataset("tips")
data_df
```

```python
#previous code 
plt.figure(figsize=(16,9))
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=True,rug=True,fit=norm,color="g",vertical=False,label="Total_bill",
             axlabel="Total_bill")
plt.title("The Bill Histogram",fontsize=20)
plt.xlabel("Total bill",fontsize=15)
plt.ylabel("Density",fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data_df["total_bill"],bins=100,hist=True,kde=True,rug=True,fit=norm,color="g",vertical=False,label="Total_bill",
             axlabel="Total_bill")
plt.title("The Bill Histogram",fontsize=20)
plt.xlabel("Total bill",fontsize=15)
plt.ylabel("Density",fontsize=15)
plt.legend()
plt.show()
```

```python
bins=[0,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data_df["total_bill"],bins=bins)
plt.xticks(bins)
plt.title("The Bill Histogram",fontsize=20)
plt.xlabel("Total bill",fontsize=15)
plt.ylabel("Density",fontsize=15)
plt.show()
```

```python
bins=[0,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(data_df["total_bill"],bins=bins,
            hist_kws={"color":"r","edgecolor":"r","linewidth":5,"linestyle":"--","alpha":0.9})
plt.xticks(bins)
plt.show()
```
