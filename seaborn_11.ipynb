```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer_data=load_breast_cancer()
cancer_data
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_data["data"],cancer_data["target"]],
                       columns=np.append(cancer_data["feature_names"], ["target"]))
cancer_df
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"])
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"],hue="target")
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"],hue="target",hue_order=[1.0,0])
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"],hue="target",hue_order=[1.0,0],
             palette="hot")
```

```python
sns.pairplot(cancer_df,hue="target",hue_order=[1.0,0],
             palette="hot",x_vars=["mean radius","mean texture"],y_vars=["mean radius"])
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"],hue="target",hue_order=[1.0,0],
             palette="hot",kind="reg")
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"],hue="target",hue_order=[1.0,0],
             palette="hot",diag_kind="hist",markers=["*",">"])
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"],hue="target",
             hue_order=[1.0,0],diag_kind="hist",markers=["*",">"],height=10,)
```

```python
sns.pairplot(cancer_df,vars=["mean radius","mean texture","mean perimeter","mean area"],hue="target",
             hue_order=[1.0,0],diag_kind="hist",markers=["*",">"],height=4)
```
