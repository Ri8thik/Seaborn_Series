```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
Gw_df=pd.read_csv("Who_is_responsible_for_global_warming.csv")
Gw_df
```

```python
new_df=Gw_df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
new_df.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df,vmin=0,vmax=21,cmap="coolwarm",annot=True)
plt.show()
```

```python
plt.figure(figsize=(16,9))
annot_kws={"fontsize":10,"fontstyle":"italic","color":"k","alpha":0.8,"rotation":'vertical',"verticalalignment":"center",
           "backgroundcolor":"w"}
sns.heatmap(new_df,vmin=0,vmax=21,cmap="coolwarm",annot=True,annot_kws=annot_kws)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df,vmin=0,vmax=21,cmap="coolwarm",annot=True,linewidth=5,linecolor="k")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df,cmap="coolwarm",annot=True,linewidth=5,linecolor="k",cbar=False)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(new_df,cmap="coolwarm",annot=True,linewidth=5,linecolor="k",cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
plt.figure(figsize=(16,9))
cbar_kws={"orientation":"horizontal",
        "shrink":1,"extend":"min","extendfrac":0.1,"ticks":np.arange(1,22),"drawedges":True}
sns.heatmap(new_df,cmap="coolwarm",annot=True,
            linewidth=5,linecolor="k",xticklabels=False,yticklabels=False,cbar_kws=cbar_kws)
plt.show()
```

```python
plt.figure(figsize=(16,9))
cbar_kws={"orientation":"horizontal",
        "shrink":1,"extend":"min","extendfrac":0.1,"ticks":np.arange(1,22),"drawedges":True}
sns.heatmap(new_df,cmap="coolwarm",annot=True,
            linewidth=5,linecolor="k",xticklabels=np.arange(1,15),yticklabels=True,cbar_kws=cbar_kws)
plt.show()
```

```python
plt.figure(figsize=(16,9))
cbar_kws={"orientation":"horizontal",
        "shrink":1,"extend":"min","extendfrac":0.1,"ticks":np.arange(1,22),"drawedges":True}
sns.heatmap(new_df,cmap="coolwarm",annot=True,
            linewidth=5,linecolor="k",xticklabels=np.arange(1,15),yticklabels=True,cbar_kws=cbar_kws)
sns.set(font_scale=20)
plt.show()
```
